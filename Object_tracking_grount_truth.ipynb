```python
from google.colab import files
import zipfile
import os

# Upload zip
uploaded = files.upload()

# Unzip
zip_name = list(uploaded.keys())[0]
with zipfile.ZipFile(zip_name, 'r') as zip_ref:
    zip_ref.extractall()

print("âœ… ZIP extracted")
```

```python
model = YOLO("yolov8n.pt")

base_dir = "/content/data"
output_base = "/content/output"
os.makedirs(output_base, exist_ok=True)

for video_folder in sorted(os.listdir(base_dir)):
    imgs_path = os.path.join(base_dir, video_folder, "imgs")

    if not os.path.isdir(imgs_path):
        continue

    print(f"â–¶ Processing {video_folder}")

    # Create output folder for this video
    video_output_dir = os.path.join(output_base, video_folder)
    os.makedirs(video_output_dir, exist_ok=True)

    gt_file = os.path.join(video_output_dir, "ground_truth.txt")

    results = model.track(
        source=imgs_path,
        conf=0.4,
        iou=0.5,
        persist=True,
        tracker="bytetrack.yaml",
        save=False,
        show=False
    )

    with open(gt_file, "w") as f:
        f.write("# frame_id x_center y_center width height\n")

        for result in results:
            frame_id = Path(result.path).stem  # 0001, 0002...

            if result.boxes is None:
                continue

            boxes = result.boxes.xywhn.cpu().numpy()

            for x, y, w, h in boxes:
                f.write(f"{frame_id} {x:.6f} {y:.6f} {w:.6f} {h:.6f}\n")

    print(f"âœ… Saved {gt_file}")

print("ðŸŽ¯ ALL videos processed successfully")
```
